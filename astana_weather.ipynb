```python
import pandas as pd
import numpy as np

df = pd.read_csv('assets/dataexport_20230319T123719.csv')
df.drop(df.index[0:9],axis = 0,inplace = True)
df = df.dropna()
df.drop(df.columns[[2,3,4]], axis = 1,inplace = True)
df.columns = df.columns.str.replace('Basel', 'Temp in C')
df.columns = df.columns.str.replace('location', 'date_time')

df


df['year'] = df['date_time'].apply(lambda x : x[:4])
df['day'] = df['date_time'].apply(lambda x : x[4:8])
df['hour'] = df['date_time'].apply(lambda x : x[9:])
df['temp'] = df['Temp in C']
df = df[~(df['day'] == '0229')]
df = df[~(df['year'] == '2023')]



df
```

```python
max22 = df[df['year'] == '2022']
min22 = df[df['year'] == '2022']
```

```python
#df_2023_mx
max12_21 = df[~(df['year'] == '2022')]
min12_21 = df[~(df['year'] == '2022')]


max12_21 = max12_21.groupby(['year','day'])['temp'].max()
min12_21 = min12_21.groupby(['year','day'])['temp'].min()

min22 = min22.groupby(['day'])['temp'].min()
max22 = max22.groupby(['day'])['temp'].min()

#df_2023

min12_21
max12_21
```

```python


max12_21 = max12_21.groupby(['day']).agg('max')
min12_21 = min12_21.groupby(['day']).agg('min')

min12_21
```

```python

merged_max = pd.merge(max12_21,max22,how = 'left', on = 'day')
merged_min = pd.merge(min12_21,min22,how = 'left', on = 'day')
merged_max
```

```python
break_max = merged_max[merged_max['temp_y'] > merged_max['temp_x']]
break_min = merged_min[merged_min['temp_y'] < merged_min['temp_x']]
break_min
```
